```python
import numpy as np

def sim(n_agents, initial_money, delta_money, n_steps):
    agents_money = np.full((n_agents), initial_money)
    for step in range(n_steps):
        winner = np.random.randint(n_agents)
        loser = np.random.randint(n_agents)
        # agents cannot run out of money, there is no 'debt'
        if agents_money[loser] >= delta_money:
            agents_money[winner] += delta_money
            agents_money[loser] -= delta_money
        
    return agents_money
```

```python
from collections import Counter
import matplotlib.pyplot as plt

agents_money = sim(1000, 10, 1, 10**7)
# Create a list of (amount, occurrences), zip it and create a list where the first
# element contains the amounts, and the second contains the occurrences (agents with that amount)
agents_money_dist = list(zip(*(sorted(Counter(agents_money).items(), reverse = True))))

plt.plot(agents_money_dist[0], agents_money_dist[1], 'bo-')
plt.xlabel('Amount of money')
plt.ylabel('Number of agents')
plt.show()
```
